Full code

````python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        
        # Period = first semester of the year 2021
        self.SetStartDate(2021,1, 1)
        self.SetEndDate(2021,7, 1)
        
        # Setting the cash to invest $500.000 USD
        self.SetCash(500000)
        
        # Selecting stock and obtaining raw data
        self.MarathonOilCorp = self.AddEquity("MRO", Resolution.Minute)
        self.MarathonOilCorp.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.DiamondBackEnergy = self.AddEquity("FANG", Resolution.Minute)
        self.DiamondBackEnergy.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.LBrands = self.AddEquity("LB", Resolution.Minute)
        self.LBrands.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.Tesla = self.AddEquity("TSLA", Resolution.Minute)
        self.Tesla.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.Apple = self.AddEquity("AAPL", Resolution.Minute)
        self.Apple.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        
    def OnData(self, data):
        if not self.Portfolio.Invested:
            self.MarketOrder("MRO", 705)
            self.Debug(str(self.Portfolio["MRO"].AveragePrice))
            
            self.MarketOrder("FANG", 400)
            self.Debug(str(self.Portfolio["FANG"].AveragePrice))
            
            self.MarketOrder("LB", 245)
            self.Debug(str(self.Portfolio["LB"].AveragePrice))
            
            self.MarketOrder("TSLA", 75)
            self.Debug(str(self.Portfolio["TSLA"].AveragePrice))
            
            self.MarketOrder("AAPL", 87)
            self.Debug(str(self.Portfolio["AAPL"].AveragePrice))
````
